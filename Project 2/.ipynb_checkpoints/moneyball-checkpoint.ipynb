```python
import sqlite3
import pandas
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

sqlite_file = 'lahman2014.sqlite'
conn = sqlite3.connect(sqlite_file)
```

Part 1: Wrangling

```python
relation = pandas.read_sql("""SELECT S.yearID, S.teamID, name,
                sum(salary) as total_payroll,
                (CAST(W as float) / CAST(G as float) * 100) as win_percent
                FROM Salaries S
                JOIN Teams T
                ON S.yearID = T.yearID
                AND S.teamID = T.teamID
                
                GROUP BY S.yearID, S.teamID
                ORDER BY win_percent
                """, conn)

relation['yearID'] = list(map(int, relation['yearID']))
relation
```

The SQL query we're using to make this relation merges data from the Teams table and the Salaries table.
To eliminate values which may be defined in one table and undefined or missing in the other table, the query only merges where a given team-year pair is defined in both tables. The values we gather are yearID, teamID, total_payroll which is the sum of all monies paid to players on a team in a given year, and the win_percent achieved by a team in the given year.
These rows are ordered by win percentage for aesthetic reasons.

<h1> Part 2: Exploratory Data Analysis

<b> Payroll Distribution

Problem 2

```python
relation_short = relation[relation['yearID'] >= 1990]
```

relation_short is a table which does not contain data from years prior to 1990

```python
teams = relation_short.groupby('teamID', as_index=False)

fig, ax = plt.subplots()
for teamID, team in teams:
    ax.scatter(team['yearID'], team['total_payroll'], marker='o', label=teamID)
    
ax.legend()
ax.set_title("Total Payroll of Teams by Year")
ax.set_xlabel('Year')
ax.set_ylabel('Total Payroll')

fig.set_figheight(10)
fig.set_figwidth(15)

plt.show()
```

We plot a scatter plot of team payrolls in a given year, nothing overly sophisticated.
In this plot, each team is given a different color, which helps distinguish trends for outlier teams, although this is not part of our analysis for this step.

Question 1

It appears that there is a moderate positive correlation between year and the total payroll of teams, and that as years have passed the average payroll has increased. It also appears that over the years, the variance of payrolls has also increased. At a glance, it does appear that as years have gone on, teams have spent more on a whole. Whether this is to improve the quality of talent, or simply in order to retain talent relative to other teams is beyond my knowledge.

Problem 3

```python
years = relation_short.groupby('yearID', as_index=False)['total_payroll'].mean()

fig, ax = plt.subplots()
for teamID, team in teams:
    ax.scatter(team['yearID'], team['total_payroll'], marker='o', label=teamID)
    
ax.plot(years['yearID'], years['total_payroll'], linewidth=10.0, color='white')
ax.plot(years['yearID'], years['total_payroll'], linewidth=5.0, color='lime')

ax.legend()
ax.set_title("Total Payroll of Teams by Year (Year by Year Mean Superimposed)")
ax.set_xlabel('Year')
ax.set_ylabel('Total Payroll')

fig.set_figheight(10)
fig.set_figwidth(15)


plt.show()
```

This is the same graph as before except I have computed the mean payroll among all teams for each year, and superimposed this data as a line plot onto the graph. This does support the assertion that as years have passed the average total payroll has increased as this line steadily increases with year.

```python
linreg = stats.linregress(relation_short['yearID'], relation_short['total_payroll'])

x_cont = np.linspace(1990, 2014, 200)
y_cont = np.array(list(map(lambda x: linreg.slope * x + linreg.intercept, x_cont)))

fig, ax = plt.subplots()
for teamID, team in teams:
    ax.scatter(team['yearID'], team['total_payroll'], marker='o', label=teamID)
    
ax.plot(x_cont, y_cont, linewidth=10.0, color='white')
ax.plot(x_cont, y_cont, linewidth=5.0, color='magenta')
    
ax.legend()
ax.set_title("Total Payroll of Teams by Year (Linear Regression of Total Payrolls Superimposed)")
ax.set_xlabel('Year')
ax.set_ylabel('Total Payroll')

fig.set_figheight(10)
fig.set_figwidth(15)

plt.show()
```

Like the previous graph, the base of this graph is our initial scatterplot. Here, I have performed a linear regression on year-total_payroll pairs. The linear function resulting from this regression is superimposed onto the scatterplot. The extremely low pvalue of this regression suggests that there is in fact a positive correlation between year and total payroll.

<B> Correlation between payroll and winning percentage

Problem 4

```python
relation_short['period'] = pandas.cut(relation_short['yearID'], [1990, 1995, 2000, 2005, 2010, 2015], include_lowest=True)
relation_short
```

Here, we are creating another column in our table. This column stores the 5 year period the given data falls into.

```python
periods = relation_short.groupby('period', as_index=False)

fig, ax = plt.subplots(5, 1)
for axi, (_, period) in zip(ax, periods):
    mean_payrolls = period.groupby('teamID', as_index=False)['total_payroll'].mean()
    mean_wins = period.groupby('teamID', as_index=False)['win_percent'].mean()
    
    for pay, w, t in zip(mean_payrolls['total_payroll'], mean_wins['win_percent'], mean_wins['teamID']):
        axi.scatter(pay, w, marker='o', label=t, s=40)
        axi.text(pay+.5, w+.25, t)
    
    linreg_period = stats.linregress(mean_payrolls['total_payroll'], mean_wins['win_percent'])
    
    x_cont = np.linspace(mean_payrolls['total_payroll'].min(), mean_payrolls['total_payroll'].max(), 200)
    y_cont = np.array(list(map(lambda x: linreg_period.slope * x + linreg_period.intercept, x_cont)))

    axi.plot(x_cont, y_cont, linewidth=10.0, color='white')
    axi.plot(x_cont, y_cont, linewidth=5.0, color='magenta')
    
    axi.set_title("Mean Payroll vs Mean Win Percent for Period " + str(period['period'].iloc[0]))
    axi.set_xlabel('Period Mean Payroll')
    axi.set_ylabel('Period Mean Win Percent')

fig.set_figheight(50)
fig.set_figwidth(15)

plt.show()
```

Using the partition from the previous cell, we group the data by 5-year period. In each period, we find the mean payroll and mean win percent for each team. For each period, we create a scatter plot which relates mean payroll and mean win percentage by team. On top of each scatter plot is superimposed a linear regression of these points. Each scatter plot displays a positive relationship between mean payroll and mean win percentage, which is supported by the linear regression.

Question 2

There appears to be a moderate positive relationship between mean payroll and mean win percentage within each 5-year period. It appears as though spread of win percentages decreases as mean payroll increases for the last three periods. This suggests that teams with higher payrolls may have experienced more consistent performance in those periods. One team which consistently outperformed the regression is the Atlanta Braves. The Oakland A's also consistently outperformed the regression which may  indicate that they had been spending their payroll judiciously all along!

<h1> Part 3: Data transformations

<b> Standardizing across years

Problem 5 

```python
year_avgs = relation_short.groupby('yearID', as_index=False)['total_payroll'].mean()
year_stddev = relation_short.groupby('yearID', as_index=False)['total_payroll'].std()

standardized = []

for _, row in relation_short.iterrows():
    avgj = year_avgs.loc[year_avgs['yearID'] == row['yearID']].total_payroll.item()
    stdj = year_stddev.loc[year_stddev['yearID'] == row['yearID']].total_payroll.item()
    
    standardized.append((row['total_payroll']-avgj)/stdj)

relation_short['std_payroll'] = standardized
relation_short
```

Here we are adding another column for the standardized payroll of a team in a given year. The standardized payroll is the number of standard deviations delta of a team's total payroll in a given year from the mean payroll of that year. To do this, we first calculate the mean and standard deviation of total payrolls for each year. Next we perform the standardization calculation on each team-year pair in the table to produce the std_payroll column.

Problem 6

```python
periods = relation_short.groupby('period', as_index=False)

fig, ax = plt.subplots(5, 1)
for axi, (_, period) in zip(ax, periods):
    mean_std_payrolls = period.groupby('teamID', as_index=False)['std_payroll'].mean()
    mean_wins = period.groupby('teamID', as_index=False)['win_percent'].mean()
    
    for pay, w, t in zip(mean_std_payrolls['std_payroll'], mean_wins['win_percent'], mean_wins['teamID']):
        axi.scatter(pay, w, marker='o', label=t, s=40)
        axi.text(pay+.01, w+.25, t)
    
    linreg_period = stats.linregress(mean_std_payrolls['std_payroll'], mean_wins['win_percent'])
    
    x_cont = np.linspace(mean_std_payrolls['std_payroll'].min(), mean_std_payrolls['std_payroll'].max(), 200)
    y_cont = np.array(list(map(lambda x: linreg_period.slope * x + linreg_period.intercept, x_cont)))

    axi.plot(x_cont, y_cont, linewidth=10.0, color='white')
    axi.plot(x_cont, y_cont, linewidth=5.0, color='magenta')
    
    axi.set_title("Mean Standard Payroll vs Mean Win Percent for Period " + str(period['period'].iloc[0]))
    axi.set_xlabel('Period Mean Standard Payroll')
    axi.set_ylabel('Period Mean Win Percent')

fig.set_figheight(50)
fig.set_figwidth(15)

plt.show()
```

Like the last set of plots, these plots relate mean standard payroll to mean win percentage in a given time period. A linear regression is overlaid on the data. As in the previous plots, we observe a moderate positive relationship between payroll and win percent.

Question 3

Over a 5-year period, the mean and standard deviation of payrolls does not vary much. Effectively, this transformation is like a shift and rescale operation. The data is shifted by the mean, and scaled by the inverse of standard deviation. The outcome is that after automatic graph limiting, which scales and shifts the data visually, the plots of the mean payroll vs win percentage and the respective plots of the mean standard payroll vs win percentage look <i> extremely </i> similar.

<b> Expected Wins

Problem 7

```python
teams = relation_short.groupby('teamID', as_index=False)

linreg = stats.linregress(relation_short['std_payroll'], relation_short['win_percent'])

x_cont = np.linspace(relation_short['std_payroll'].min(), relation_short['std_payroll'].max(), 200)
y_cont = np.array(list(map(lambda x: linreg.slope * x + linreg.intercept, x_cont)))

fig, ax = plt.subplots()
for teamID, team in teams:
    ax.scatter(team['std_payroll'], team['win_percent'], marker='o', label=teamID)
    
ax.plot(x_cont, y_cont, linewidth=10.0, color='white')
ax.plot(x_cont, y_cont, linewidth=5.0, color='magenta')
    
ax.legend()
ax.set_title("Standard Payroll vs Win Percentage (Linear Regression of Win Percentage Superimposed)")
ax.set_xlabel('Standard Payroll')
ax.set_ylabel('Win Percentage')

fig.set_figheight(10)
fig.set_figwidth(15)

plt.show()
```

Here we are creating a scatter plot standard payroll vs win percentage. Data for each team is grouped and assigned a singular color for aesthetic reasons. This scatter plot suggests a moderate positive relationship between standard payroll and win percentage. A linear regression is superimposed onto this scatter plot which supports the conclusion that there is a positive relationship between these variables.

<b> Spending Efficiency

Problem 8

```python
expectation = []

for _, row in relation_short.iterrows():
    expectation.append(50 + 2.5*row['std_payroll'])

relation_short['expected_percent'] = expectation
relation_short
```

Here we are adding a new column to the table. This column is the expected win percentage of a team in a given year. This is calculated as a linear function of that team's standardized payroll for that year.

```python
relation_short['efficiency'] = relation_short['win_percent'] - relation_short['expected_percent']
relation_short
```

Here we are adding yet another column to the table. The efficiency column represents the percent delta between a team's actual win percentage and their expected win percentage in a given year. This is a element-wise column operation. A positive efficiency means that the team outperformed the expectations of this model in that year; a negative efficiency means that the team underperformed by the same metric.

```python
fig, ax = plt.subplots(5, 1)
for axi, teamID in zip(ax, ['OAK', 'BOS', 'NYA', 'ATL', 'TBA']):
    team_frame = relation_short.loc[relation_short['teamID'] == teamID].sort_values('yearID')
    efficiencies = team_frame['efficiency']
    years = team_frame['yearID']
    axi.plot(years, efficiencies, marker='o', label=teamID)
    axi.plot(np.linspace(1990, 2015, 200), np.zeros(200), color="red")
    
    axi.legend()
    axi.set_title("Efficiency of " + team_frame['name'].iloc[0] + " by Year")
    axi.set_xlabel('Year')
    axi.set_ylabel('Efficiency')

fig.set_figheight(50)
fig.set_figwidth(15)

plt.show()
```

These are our final plots. I find them the most compelling because each one conveys a narrative of potential realized and maximized, and of money wasted. For each of the 5 teams selected, a line plot is created of that team's efficiency from 1990 to 2015. Superimposed onto this is x-axis. Points above the x-axis are years where that team beat our model, and points below are the opposite.
